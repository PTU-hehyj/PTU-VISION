```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # ColabÏö© Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†• Ìï®Ïàò

# YOLO Î™®Îç∏ Î°úÎìú
model = YOLO("yolo11x-seg.pt")

# Ïó¨Îü¨ Ïù¥ÎØ∏ÏßÄ ÏòàÏ∏° ÏàòÌñâ
image_paths = ["./input.jpg", "./input2.jpg", "./input3.JPG", "./input4.jpg"]
results = model(image_paths)

# ÏòàÏ∏° Í≤∞Í≥º ÏãúÍ∞ÅÌôî Î∞è Î¶¨ÏÇ¨Ïù¥Ï¶à
for idx, r in enumerate(results):
    image = r.plot()  # ÏòàÏ∏°Îêú Í∞ùÏ≤¥Í∞Ä Ìè¨Ìï®Îêú Ïù¥ÎØ∏ÏßÄ
    resized_image = cv2.resize(image, (640, 640))  # 640x640 ÌÅ¨Í∏∞Î°ú Ï°∞Ï†ï

    # ColabÏóêÏÑú Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
    cv2_imshow(resized_image)

    # Í≤∞Í≥º Ï†ÄÏû• (Í∞Å ÌååÏùºÏùÑ output_1.jpg, output_2.jpg ... ÌòïÏãùÏúºÎ°ú Ï†ÄÏû•)
    cv2.imwrite(f"output_{idx+1}.jpg", resized_image)
```
