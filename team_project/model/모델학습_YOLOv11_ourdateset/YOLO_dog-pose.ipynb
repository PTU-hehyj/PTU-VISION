```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Colabìš© ì´ë¯¸ì§€ ì¶œë ¥ í•¨ìˆ˜

# YOLO ëª¨ë¸ ë¡œë“œ
model = YOLO("best.pt")

# ì—¬ëŸ¬ ì´ë¯¸ì§€ ì˜ˆì¸¡ ìˆ˜í–‰
image_paths = ["./input.jpg", "./input2.jpg", "./input3.jpg", "./input4.jpeg"]
results = model(image_paths, conf=0.1)

# ì˜ˆì¸¡ ê²°ê³¼ ì‹œê°í™” ë° ë¦¬ì‚¬ì´ì¦ˆ
for idx, r in enumerate(results):
    image = r.plot()  # ì˜ˆì¸¡ëœ ê°ì²´ê°€ í¬í•¨ëœ ì´ë¯¸ì§€
    resized_image = cv2.resize(image, (640, 640))  # 640x640 í¬ê¸°ë¡œ ì¡°ì •

    # Colabì—ì„œ ì´ë¯¸ì§€ ì¶œë ¥
    cv2_imshow(resized_image)

    # ê²°ê³¼ ì €ìž¥ (ê° íŒŒì¼ì„ output_1.jpg, output_2.jpg ... í˜•ì‹ìœ¼ë¡œ ì €ìž¥)
    cv2.imwrite(f"output_{idx+1}.jpg", resized_image)
```
