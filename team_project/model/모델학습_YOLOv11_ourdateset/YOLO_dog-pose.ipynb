```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Colab용 이미지 출력 함수

# YOLO 모델 로드
model = YOLO("best.pt")

# 여러 이미지 예측 수행
image_paths = ["./input.jpg", "./input2.jpg", "./input3.jpg", "./input4.jpeg"]
results = model(image_paths, conf=0.1)

# 예측 결과 시각화 및 리사이즈
for idx, r in enumerate(results):
    image = r.plot()  # 예측된 객체가 포함된 이미지
    resized_image = cv2.resize(image, (640, 640))  # 640x640 크기로 조정

    # Colab에서 이미지 출력
    cv2_imshow(resized_image)

    # 결과 저장 (각 파일을 output_1.jpg, output_2.jpg ... 형식으로 저장)
    cv2.imwrite(f"output_{idx+1}.jpg", resized_image)
```
